```python
# damages is object type == not good
# change to float 

def type_conversion(lst):
    conversion = {"M": 1000000, "B": 1000000000}
    new_lst = []
    for i in lst:
        if i == 'Damages not recorded':
            new_lst.append(float(0))
        elif i[-1] == 'M':
            new_lst.append(float(i.strip('M')) * conversion['M'])
        elif i[-1] == 'B':
            new_lst.append(float(i.strip('B')) * conversion['B'])
    return new_lst
            
```

```python
data['damages'] = type_conversion(data['damages'])
```

```python
# pandas cant iterate inside every list in the column
# lets fix that

# https://towardsdatascience.com/dealing-with-list-values-in-pandas-dataframes-a177e534f173

def to_1D(series):
    return pd.Series([x for _list in series for x in _list])
```

```python
to_1D(data['Areas Affected']).value_counts()
```

## Create a simple plot

```python
idx = to_1D(data['Areas Affected']).value_counts().index  # index so i can plot x label
numbers = to_1D(data['Areas Affected']).value_counts().values  # values, y label
```

```python
# temporary style
with plt.style.context('seaborn'):
    fig, ax = plt.subplots(figsize=(18,6))

ax.bar(idx, numbers)


fig.autofmt_xdate()  # Use to fix theoverlap in x axis

plt.tight_layout()
plt.show()
```

```python
def mort_scale(l):
    mortality_scale = {0: 0,
                   1: 100,
                   2: 500,
                   3: 1000,
                   4: 10000}
    mortality = []
    for i in l:
        if i == 0:
            mortality.append(0)
        elif i <= 100:
            mortality.append(1)
        elif i <= 500:
            mortality.append(2)
        elif i <= 1000:
            mortality.append(3)
        elif i <= 10000:
            mortality.append(4)
        else:
            mortality.append(5)
            
    return mortality
```

```python
data['mortality_scale'] = mort_scale(data['deaths'])
```

```python
def damage(lst):
    damage_scale = {0: 0,
                1: 100000000,
                2: 1000000000,
                3: 10000000000,
                4: 50000000000}
    damage = []
    for i in lst:
        if i == 0:
            damage.append(0)
        elif i <= 100000000:
            damage.append(1)
        elif i <= 1000000000:
            damage.append(2)
        elif i <= 10000000000:
            damage.append(3)
        elif i <= 50000000000:
            damage.append(4)
        else:
            damage.append(5)
            
    return damage
```

```python
data['damage_scale'] = damage(data['damages'])
```
